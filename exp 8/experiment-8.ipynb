```python
import cv2
import numpy as np  # for shapes

def convert_to_grayscale(image_path):
    # Load the image
    image = cv2.imread(image_path)

    if image is None:
        print("Error: Image not found at", image_path)
        return

    # Convert to grayscale
    grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Display the original and grayscale images
    print("Original Image:")
    cv2.imshow("Original Image", image)

    print("Grayscale Image:")
    cv2.imshow("Grayscale Image", grayscale_image)

    cv2.waitKey(0)
    cv2.destroyAllWindows()

def convert_to_bgr(image_path, output_path):
    # Load the image
    image = cv2.imread(image_path)

    if image is None:
        print("Error: Image not found at", image_path)
        return

    # No need to convert to BGR again because imread loads in BGR by default
    bgr_image = image.copy()

    # Display the original and BGR images
    print("Original Image:")
    cv2.imshow("Original Image", image)

    print("BGR Image:")
    cv2.imshow("BGR Image", bgr_image)

    cv2.waitKey(0)
    cv2.destroyAllWindows()

    # Save the BGR image
    cv2.imwrite(output_path, bgr_image)
    print("BGR image saved as", output_path)

def convert_bgr_to_rgb(image_path):
    # Load the image
    image = cv2.imread(image_path)

    if image is None:
        print("Error: Image not found at", image_path)
        return

    # Convert to RGB
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    print("BGR Image:")
    cv2.imshow("BGR Image", image)

    print("RGB Image:")
    cv2.imshow("RGB Image", rgb_image)

    cv2.waitKey(0)
    cv2.destroyAllWindows()

# --------- MAIN PROGRAM ---------
image_path = "OIP.jpeg"
output_path = "output_bgr.jpg"

convert_to_grayscale(image_path)
convert_to_bgr(image_path, output_path)
convert_bgr_to_rgb(output_path)
```
